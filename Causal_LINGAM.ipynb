```python
# ============================================================
# BLOCK 3: PREPROCESS DATA FOR LINGAM
# ============================================================

print("\n" + "="*60)
print("PREPROCESSING FOR LINGAM")
print("="*60)

# Store original dataframe
df_original = df.copy()

# Get column names
columns = df.columns.tolist()

# Handle missing values
if df.isnull().sum().sum() > 0:
    print(f"\nDropping {df.isnull().sum().sum()} missing values...")
    df = df.dropna()
    print(f"Remaining rows: {len(df)}")

# Identify column types
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

print(f"\nCategorical columns: {categorical_cols}")
print(f"Numeric columns: {numeric_cols}")

# Encode categorical columns for LiNGAM
df_encoded = df.copy()
encoders = {}

for col in categorical_cols:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df[col].astype(str))
    encoders[col] = le
    print(f"Encoded '{col}': {len(le.classes_)} unique values")

# Update columns list
columns = df_encoded.columns.tolist()

print(f"\nPreprocessed data shape: {df_encoded.shape}")
print(f"\nPreprocessed data preview:")
print(df_encoded.head())
```

```python
# ============================================================
# BLOCK 4: LEARN CAUSAL STRUCTURE WITH LINGAM
# ============================================================

print("\n" + "="*60)
print("LINGAM CAUSAL DISCOVERY")
print("="*60)

# Fit LiNGAM model
lingam_model = DirectLiNGAM(random_state=42)
lingam_model.fit(df_encoded.values)

# Get adjacency matrix (causal coefficients)
adj_matrix = lingam_model.adjacency_matrix_

# Get causal order
causal_order = lingam_model.causal_order_
causal_order_names = [columns[i] for i in causal_order]

print(f"\nCausal Order Discovered:")
for i, var in enumerate(causal_order_names):
    print(f"  {i+1}. {var}")

print(f"\nAdjacency Matrix (Causal Coefficients):")
adj_df = pd.DataFrame(adj_matrix, index=columns, columns=columns).round(3)
print(adj_df)

print(f"\nReading: adj_matrix[i,j] ≠ 0 means column j CAUSES row i")
```

```python
# ============================================================
# BLOCK 5: EXTRACT CAUSAL EDGES
# ============================================================

print("\n" + "="*60)
print("CAUSAL GRAPH EXTRACTION")
print("="*60)

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# MODIFY THIS: Adjust threshold if needed
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

THRESHOLD = 0.01  # Minimum coefficient to consider as edge

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

edges = []
edge_weights = {}

print(f"\nDiscovered causal relationships (threshold > {THRESHOLD}):")
print("-"*50)

for i in range(len(columns)):
    for j in range(len(columns)):
        if abs(adj_matrix[i, j]) > THRESHOLD:
            from_var = columns[j]
            to_var = columns[i]
            weight = adj_matrix[i, j]
            edges.append((from_var, to_var))
            edge_weights[(from_var, to_var)] = weight
            print(f"  {from_var} → {to_var} (coefficient: {weight:.4f})")

print(f"\nTotal edges discovered: {len(edges)}")

if len(edges) == 0:
    print("\n⚠ WARNING: No edges found! Try lowering the THRESHOLD.")
```

```python
# ============================================================
# BLOCK 6: VISUALIZE CAUSAL GRAPH
# ============================================================

print("\n" + "="*60)
print("VISUALIZE CAUSAL GRAPH")
print("="*60)

try:
    import networkx as nx

    # Create directed graph
    G = nx.DiGraph()
    G.add_nodes_from(columns)
    G.add_edges_from(edges)

    # Plot
    plt.figure(figsize=(12, 8))
    pos = nx.spring_layout(G, k=2, iterations=50, seed=42)

    # Draw nodes
    nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=2000, alpha=0.9)
    nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')

    # Draw edges with weights
    nx.draw_networkx_edges(G, pos, edge_color='gray', arrows=True,
                           arrowsize=20, connectionstyle='arc3,rad=0.1')

    # Add edge labels
    edge_labels = {k: f"{v:.2f}" for k, v in edge_weights.items()}
    nx.draw_networkx_edge_labels(G, pos, edge_labels, font_size=8)

    plt.title("Causal Graph (LiNGAM)", fontsize=14)
    plt.axis('off')
    plt.tight_layout()
    plt.savefig('causal_graph.png', dpi=150, bbox_inches='tight')
    plt.show()

    print("✓ Saved: causal_graph.png")

except ImportError:
    print("⚠ networkx not installed. Skipping visualization.")
except Exception as e:
    print(f"⚠ Could not visualize: {e}")
```

```python
# ============================================================
# BLOCK 7: SAVE LINGAM STRUCTURE
# ============================================================

print("\n" + "="*60)
print("SAVE CAUSAL STRUCTURE")
print("="*60)

# Create structure dictionary
causal_structure = {
    'columns': columns,
    'causal_order': causal_order_names,
    'edges': edges,
    'edge_weights': {str(k): float(v) for k, v in edge_weights.items()},
    'adjacency_matrix': adj_matrix.tolist(),
    'threshold_used': THRESHOLD
}

# Save as JSON
with open('causal_structure.json', 'w') as f:
    json.dump(causal_structure, f, indent=2)
print("✓ Saved: causal_structure.json")

# Save as pickle (includes model object)
with open('causal_structure.pkl', 'wb') as f:
    pickle.dump({
        'columns': columns,
        'causal_order': causal_order_names,
        'edges': edges,
        'edge_weights': edge_weights,
        'adjacency_matrix': adj_matrix,
        'lingam_model': lingam_model,
        'encoders': encoders
    }, f)
print("✓ Saved: causal_structure.pkl")

# Save adjacency matrix as CSV
adj_df.to_csv('adjacency_matrix.csv')
print("✓ Saved: adjacency_matrix.csv")
```

```python
# ============================================================
# BLOCK 8: IDENTIFY CONFOUNDERS
# ============================================================

print("\n" + "="*60)
print("IDENTIFY CONFOUNDERS")
print("="*60)

def find_confounders(edges, treatment, outcome):
    """
    Find confounders for a treatment-outcome pair.
    A confounder causes BOTH treatment and outcome.
    """
    causes_treatment = set()
    causes_outcome = set()

    for (from_var, to_var) in edges:
        if to_var == treatment:
            causes_treatment.add(from_var)
        if to_var == outcome:
            causes_outcome.add(from_var)

    confounders = causes_treatment.intersection(causes_outcome)
    return list(confounders)

# Show confounders for each possible treatment-outcome pair
print("\nConfounder analysis for all variable pairs:")
print("-"*60)

confounder_dict = {}

for treatment in columns:
    for outcome in columns:
        if treatment != outcome:
            confounders = find_confounders(edges, treatment, outcome)
            if confounders:
                key = f"{treatment} → {outcome}"
                confounder_dict[key] = confounders
                print(f"\n{treatment} → {outcome}:")
                print(f"  Confounders: {confounders}")

if not confounder_dict:
    print("\nNo confounders found for any variable pairs.")
```

```python
# ============================================================
# BLOCK 9: DISCRETIZE DATA FOR BAYESIAN NETWORK
# ============================================================

print("\n" + "="*60)
print("DISCRETIZE DATA FOR BAYESIAN NETWORK")
print("="*60)

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# MODIFY THIS: Adjust number of bins if needed
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

N_BINS = 3  # Number of bins for discretization

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

df_discrete = pd.DataFrame()

for col in columns:
    unique_vals = df_encoded[col].nunique()

    # If already few unique values, treat as categorical
    if unique_vals <= N_BINS + 2:
        df_discrete[col] = df_encoded[col].astype(str)
        print(f"{col}: Kept as is ({unique_vals} unique values)")
    else:
        # Discretize continuous variables
        try:
            discretizer = KBinsDiscretizer(n_bins=N_BINS, encode='ordinal', strategy='quantile')
            df_discrete[col] = discretizer.fit_transform(df_encoded[[col]]).flatten().astype(int).astype(str)
            print(f"{col}: Discretized into {N_BINS} bins")
        except Exception as e:
            df_discrete[col] = df_encoded[col].astype(str)
            print(f"{col}: Could not discretize ({e}), kept as is")

print(f"\nDiscretized data shape: {df_discrete.shape}")
print(f"\nValue counts per column:")
for col in columns[:5]:  # Show first 5 columns
    print(f"\n{col}:")
    print(df_discrete[col].value_counts())
```

```python
# ============================================================
# BLOCK 10: CREATE BAYESIAN NETWORK
# ============================================================

print("\n" + "="*60)
print("CREATE BAYESIAN NETWORK")
print("="*60)

# Check if we have edges
if len(edges) == 0:
    print("⚠ ERROR: No edges found from LiNGAM. Cannot create Bayesian Network.")
    print("  Try lowering the THRESHOLD in BLOCK 5.")
else:
    # Create Bayesian Network using edges from LiNGAM
    print(f"Using {len(edges)} edges from LiNGAM")

    try:
        bn_model = BayesianNetwork(edges)

        print(f"\nBayesian Network structure:")
        print(f"  Nodes: {list(bn_model.nodes())}")
        print(f"  Edges: {list(bn_model.edges())}")

        # Fit the model with data
        bn_model.fit(df_discrete, estimator=MaximumLikelihoodEstimator)

        print("\n✓ Bayesian Network created and fitted!")

        # Show CPD summary
        print("\nConditional Probability Distributions (CPDs):")
        for cpd in bn_model.get_cpds():
            print(f"  {cpd.variable}: parents = {list(cpd.get_evidence()) if cpd.get_evidence() else 'None (root node)'}")

        BN_CREATED = True

    except Exception as e:
        print(f"\n⚠ ERROR creating Bayesian Network: {e}")
        print("\nPossible issues:")
        print("  - Graph may have cycles (LiNGAM should prevent this)")
        print("  - Too few samples for some variable combinations")
        print("  - Try adjusting N_BINS in BLOCK 9")
        BN_CREATED = False
```
